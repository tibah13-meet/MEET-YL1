```python
import csv
f = open("environmental.csv", "rb")
r = csv.reader(f)
rownum = 0
radiation = []
temperature = []

for i in r:
    if rownum != 0:
        z = map(float, i)
        radiation.append(z[1])
        temperature.append(z[2])
    rownum +=1
    
f.close()

from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Radiation and Temperature :)')
ax.set_xlabel('Radiation')
ax.set_ylabel('Temperature')
bar(radiation, temperature)
```

This graph shows connection or relation between Radiation and Temperature
